# Slater Determinants
The Slater determinant for free fermions is
$$
    \abs{\det M}^2 = \frac{1}{L^{N_p}} \prod_{1 \leq n < m \leq L} \abs{2 \sin \left( \frac{\pi}{L} (m - n)\right)}^2,
$$
where $L$ is the length of the chain, $N_p$ the number of particles and $n, m$ are the positions of the fermions.
We are interested in $\log \abs{\det M}^2$.

For a staggered configuration, where $n, m \in \{1, 3, \dots\}$ we have to distinguish even and odd $L$.



## Even case
We have $L = 2 \ell$ and $N_p = \ell$.
The maximal configuration is given by the positions
$$
n, m \in \{1, 3, L - 1\}.
$$
The product inside the Slater determinant is over all the possible distances between a pair of particles.
These distances can be indexed with a single integer $j$, such that $m - n = 2j$.
A distance of $m - n = 2j$ appears $\ell - j$ times in the product.

With these knowledge the Slater determinant can be simplified into the form:
$$
\slater{even} = \frac{1}{(2\ell)^\ell} \abs{ \prod_{j = 1}^{\ell-1} 2 \sineven{j}^{\ell - j} }.
$$
Let $\normconst{even} = (2\ell)^{-\ell}$ be the normalization constant.
The log of the determinant can be written as:
$$
\log \slater{even} = - \log \normconst{even} + 2 \bigsum{even}(\ell)
$$
where $\bigsum{even}(\ell)$ is the sum
$$
\bigsum{even}(\ell) = \sum_{j = 1}^{\ell - 1} (\ell - j) \func{even}(j, \ell), \qquad
\func{even}(j, \ell) = \log \left( 2 \sineven{j} \right).
$$

Through some manipulation $\bigsum{even}$ can be brought in the form
$$
\bigsum{even}(\ell) = \ell\, \smallsum{even}(\ell) - \delta_{\ell}, \qquad
\smallsum{even}(\ell) = \sum_{j = 1}^{\ell/2} \func{even}(j, \ell)
$$
Notice that the last sum is just of half of the terms.
When $\ell$ is odd, by $\ell / 2$ we mean the integer part of $\ell / 2$.
The extra term $\delta_{\ell}$ is:
$$
\delta_{\ell} = \frac{\ell}{2} \func{even} \left( \frac{\ell}{2}, \ell \right) = \frac{\ell}{2} \log 2
$$
only when $\ell$ is even. Zero otherwise.
This appears due to the parity of $\ell$ (because we have split the sum).

We want to compare $\bigsum{even}(\ell)$ and $\smallsum{even}(\ell)$ to the odd case.
In the code we refer to $\bigsum{}$ before and after the decomposition as "**full sum**" and "**simplified sum**", respectevely.



## Odd case
We have $L = 2 \ell + 1$ and $N_p = \ell + 1$.
In this case the maximal configuration is given by the positions
$$
n, m \in \{1,\, 3,\, \dots,\, 2 \ell + 1\}
$$
Similarly to the even case, all the possible distances can be indexed with a single integer $j$, such that $m - n = 2j$.
But in this case a distance of $m - n = 2j$ appears $l - j + 1$ times.

Therefore, the Slater determinant can be simplified in the form:
$$
\slater{odd} = \frac{1}{(2\ell + 1)^{\ell+1}} \abs{\prod_{j = 1}^{\ell} 2 \sinodd{j}^{\ell - j + 1}}
$$
Let $\normconst{odd} = (2\ell + 1)^{-\ell - 1}$ be the normalization constant.
We can recast the log of the determinant similarly to the even case:
$$
\log \slater{odd} = - \log \normconst{odd} + 2 \bigsum{odd}(\ell)
$$
Now the full sum $\bigsum{odd}$ depends on three quantities:
$$
\bigsum{odd}(\ell) = \sum_{j = 1}^{\ell - 1} (\ell - j + 1) \func{odd}(j, \ell), \qquad
\func{odd}(j, \ell) = \log \left( 2 \sinodd{j} \right),
$$
Which looks similar to the even case with some modifications.

Through some manipulation the full sum $\bigsum{odd}$ can be brought in the form
$$
\bigsum{odd}(\ell) = \ell\, \smallsum{odd}(\ell) + \Delta(\ell) +  \overline{\delta}_{\ell}, \qquad
$$
The phi term has a similar definition,
$$
\smallsum{odd}(\ell) = \sum_{j = 1}^{\ell/2} \func{odd}(j, \ell),
$$
but a new term, a deviation, appears in the full sum:
$$
\Delta(\ell) = \sum_{j = 1}^{\ell / 2} j \left( \log \sinodd{(j - \onehalf)} - \log \sinodd{j} \right)
$$
The term $\overline{\delta}_{\ell}$ is an analogue of $\delta_{\ell}$:
$$
\overline{\delta}_{\ell} = \frac{\ell + 1}{2} \func{odd} \left( \frac{\ell + 1}{2}, \ell \right)
$$
only when $\ell$ is odd.
For $\ell \gg 1$ we have $\overline{\delta}_{\ell} \approx \left( \frac{\ell + 1}{2} \right) \log 2$.


## Summary
**Log of the Slater determinant**
$$
\log \slater{even} = - \log \normconst{even} + 2 \bigsum{even}(\ell)
$$
$$
\log \slater{odd} = - \log \normconst{odd} + 2 \bigsum{odd}(\ell)
$$

**Summand function**
$$
\func{even}(j, \ell) = \log \left( 2 \sineven{j} \right), \qquad
\func{odd}(j, \ell) =  \log \left( 2 \sinodd{j} \right)
$$

**Full sums (lhs) and simplified sums (rhs)**
$$
\bigsum{even}(\ell) = \sum_{j = 1}^{\ell - 1} (\ell - j) \func{even}(j, \ell) = \ell \smallsum{even}(\ell) + \delta_{\ell}
$$
$$
\bigsum{odd}(\ell) = \sum_{j = 1}^{\ell} (\ell - j + 1) \func{odd}(j, \ell) = (\ell + 1) \smallsum{odd}(\ell) + \Delta(\ell) + \delta_{\ell}
$$

**Phi terms**
$$
\smallsum{even} = \sum_{j = 1}^{\ell / 2} \func{even}(j, \ell), \qquad
\smallsum{odd} = \sum_{j = 1}^{\ell / 2} \func{odd}(j, \ell)
$$

**Deviation Delta term**
$$
\Delta(\ell) = \sum_{j = 1}^{\ell / 2} j \left( \log \sinodd{(j - \onehalf)} - \log \sinodd{j} \right)
$$

**Parity correction to the sums**

Even case:
$$
\delta_{\ell} = \frac{\ell}{2} \func{even} \left( \frac{\ell}{2}, \ell \right) = \frac{\ell}{2} \log 2
$$
only for $\ell$ even

Odd case:
$$
\overline{\delta}_{\ell} = \frac{\ell + 1}{2} \func{odd} \left( \frac{\ell + 1}{2}, \ell \right)
$$
only for $\ell$ odd

# Summary of the Results


Taking the difference between the logs of the Slater determinants we have
$$
\log \slater{odd} - \log \slater{even} =
- \log \frac{\normconst{odd}}{\normconst{even}} +
+ 2 \left(
    \ell ( \smallsum{odd} - \smallsum{even}) + \smallsum{odd} + \Delta + (\delta_{\ell} - \overline{\delta}_{\ell})
    \right)
$$
The last term yields $\delta_{\ell} - \overline{\delta}_{\ell} \sim (\ell / 2) \log 2$ for all $\ell$.

From the numerics we get the behaviour of the other terms:
$$
\log \frac{\normconst{odd}}{\normconst{even}} \sim \log (\ell)
$$

$$
\smallsum{odd} - \smallsum{even} \sim \frac{1}{4 \ell} + c, \qquad
c \sim -0.173 \sim - \frac{1}{4} \log 2
$$

$$
\smallsum{odd} \sim \frac{1}{2} \log (\ell)
$$

$$
\Delta \sim a \ell + b \log(\ell), \qquad
a \sim c, \quad
b \sim \frac{1}{8}
$$

With these values, a lot of terms cancel out and we obtain the final result:
$$
 - \log \slater{odd} + \log \slater{even} \sim \frac{1}{4} \log(\ell) + \mathrm{const}
$$
Which is compatible with results below of the inf. Renyi Shannon entropy
$$ - \log p_{\text{max}} = - \log | \det M |^2 $$
