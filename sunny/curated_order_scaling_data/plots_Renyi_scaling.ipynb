```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc("text", usetex=True)
plt.rc("text.latex", preamble=r"\usepackage{physics} \usepackage{mathtools}")
plt.rc("font", family="serif", size=28, weight="bold")
```

```python
N_odd_range = [11, 25]
renyi_entropy_orders = np.linspace(1, 11, 21)
global_neg = False
delta = 0
print_mode = True
initials = [0.1, 0.1, 0.1]
```

```python
import os

data_filepath = f"Renyi_a_scaling_with_order_{renyi_entropy_orders[0]}_{renyi_entropy_orders[-1]}_globalneg_{global_neg}_N_{N_odd_range[0]}_{N_odd_range[-1] - 2}.npy"
remove_last = 0

data = np.load(data_filepath)

sl = slice(None, -remove_last if remove_last > 0 else None, None)
fig, ax = plt.subplots(1, 1, figsize=[12, 8])
yplot = data[1]
plt.plot(renyi_entropy_orders[sl], yplot[sl], "+", color="tab:orange", markersize=20, markeredgewidth=3)
plt.xlabel(r"$n$")
plt.ylabel(r"$a$")

plt.grid(color="gray", linestyle="dashdot", linewidth=1.6)
plt.tight_layout()
plt.savefig(f"Renyi_a_scaling_with_order_{renyi_entropy_orders[0]}_{renyi_entropy_orders[-1]}_globalneg_{global_neg}_N_{N_odd_range[0]}_{N_odd_range[-1] - 2}.pdf", dpi=300, facecolor="none")
```
