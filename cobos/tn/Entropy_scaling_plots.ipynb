## Plot settings

```python
N_sim_range = [3, 23]
N_plot_range = [9, 23]
jump2 = True
delta_range = [-1, 1]
ndeltas = 21
global_neg = True
maxbdim = 256
nsweeps = 10
cutoff = 1e-12
regression_initials = [0.1, 0.1, 0.1]

probabilies_folder = "Results/"
plots_folder = ""
```

## Plots

```python
from plotting import min_entropy_delta_L_scaling_plots
import numpy as np
import os

probabilities_filename = f"heisemberg_gs_energy_maxp_N_{N_sim_range[0]}_{N_sim_range[1]}_delta_{delta_range[0]}_{delta_range[1]}_ndeltas_{ndeltas}_globalneg_{"true" if global_neg else "false"}_maxbdim_{maxbdim}_sweeps_{nsweeps}_cutoff_{cutoff:.2E}.npy"
probabilities_filepath = os.path.join(probabilies_folder, probabilities_filename)
energies, max_probabilities = np.load(probabilities_filepath)
plot_filename = f"min_entropy_scaling_N_{N_sim_range[0]}_{N_sim_range[1]}_delta_{delta_range[0]}_{delta_range[1]}_ndeltas_{ndeltas}_globalneg_{"true" if global_neg else "false"}_maxbdim_{maxbdim}_sweeps_{nsweeps}_cutoff_{cutoff:.2E}.pdf"
plot_filepath = os.path.join(plots_folder, plot_filename)

plot_slice = slice(int(np.where(np.equal(N_plot_range[0], np.arange(N_sim_range[0], N_sim_range[1]+1)))[0]), int(np.where(np.equal(N_plot_range[1], np.arange(N_sim_range[0], N_sim_range[1]+1)))[0]+1), 1 + jump2)
L_arr = np.arange(N_plot_range[0], N_plot_range[1]+1, 1 + jump2)
min_entropy_delta_L_scaling_plots(max_probabilities[plot_slice, :], L_arr, delta_range, regression_initials, plot_filepath)
```
